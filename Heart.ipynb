```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy as np
import math
import scipy
```

```python
data= [[-2], [-1.75], [-1.5], [-1.25], [-1], [-0.75], [-0.5], [0], [0.5], [0.75], [1], [1.25], [1.5], [1.75], [2]]
data
```

```python
df = pd.DataFrame(data, columns = ['x'])
df
```

```python
df['|x|'] = np.abs(df['x'])
df
```

```python
df['y_1'] = np.sqrt(1 - (df['|x|']-1)**2)
df
```

```python
df['y_2'] = -3*np.sqrt((1-np.sqrt(df['|x|'])/np.sqrt(2)))
df
```

```python
plt.plot(df['x'], df['y_1'])
plt.plot(df['x'], df['y_2'])
plt.fill(df['x'], df['y_1'], "r",df['x'], df['y_2'], "r")
```
